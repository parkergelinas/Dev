```python
import sqlalchemy
import pymysql
import ta
import pandas as pd
import numpy as np 
```

```python
pymysql.install_as_MySQLdb()
```

```python
engine = sqlalchemy.create_engine('mysql://root:1234@localhost:3306')
```

```python
def gettables(index):
    query = f"""SELECT table_name FROM information_schema.tables
    WHERE table_schema = '{index}'"""
    df = pd.read_sql(query, engine)
    df['Schema'] = index
    return df
```

```python
gettables('Nifty50')
```

```python
nifty = gettables('Nifty50')
```

```python
def getprices(which):
    prices = []
    for table, schema in zip(which.TABLE_NAME, which.Schema):
        sql = schema+'.'+f'`{table}`'
        prices.append(pd.read_sql(f"SELECT Date, Close FROM {sql}",engine))
    return prices
```

```python
getprices(nifty)
```
